**2. Import library dan membaca dataframe (df) excel**

```python
import pandas as pd

# kalau Excel
df = pd.read_excel("Data_Tanaman_Padi_Sumatera_version_1.xlsx")

```

**3. Menampilkan Datasheet kotor**

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
print(df)
```

**3. Data Cleansing-Standarisasi Produksi**

```python
df['Produksi'] = df['Produksi'].astype(float).astype(int)
```

**4. Data Cleansing-Standarisasi Luas panen**

```python
df['Luas Panen'] = df['Luas Panen'].astype(float).astype(int)
```

**5. Data Cleansing-Standarisasi Curah hujan**

```python
df['Curah hujan'] = df['Curah hujan'].astype(float).astype(int)
```

**6. Data Cleansing-Standarisasi Kelemban**

```python
df['Kelembapan'] = df['Kelembapan'].round().astype(int)
```

**7. Data Cleansing-Standarisasi Suhu rata-rata**

```python
df['Suhu rata-rata'] = df['Suhu rata-rata'].round().astype(int)
```

**8. Menghapus Data Duplikate**

```python
df = df.drop_duplicates()
```

**9. Menambahkan angka 0 pada produksi agar sama jumlah digit angka**

```python
df['Produksi'] = df['Produksi'].astype(str).str.ljust(7, '0')
```

**10. Menambahkan angka 0 pada Luas Panen**

```python
df['Luas Panen'] = df['Luas Panen'].astype(str).str.ljust(6, '0')
```

**11. Menambahkan angka 0 pada Curah hujan**

```python
df['Curah hujan'] = df['Curah hujan'].astype(str).str.ljust(4, '0')
```
